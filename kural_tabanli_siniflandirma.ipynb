```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("C:/Users/Emre Bilir/Desktop/DSBootcamp/Kural Tabanlı Sınıflandırma/persona.csv")
df.head()
```

```python
agg_df = df.groupby(["COUNTRY","SOURCE","AGE","SEX"]).agg({"PRICE":"mean"}).sort_values(by="PRICE", ascending=False)
agg_df
```

```python
agg_df.reset_index(inplace=True)
agg_df
```

```python
agg_df["AGE_CAT"] = pd.cut(agg_df["AGE"], bins=[0,18,23,30,40,70], labels=['0_18', '19_23', '24_30', '31_40', '41_70'])
agg_df
```

```python
agg_df["customers_level_based"] = agg_df[["COUNTRY", "SOURCE", "SEX", "AGE_cat"]].agg(lambda x: '_'.join(x).upper(), axis=1)
agg_df
```

```python
agg_df = agg_df[["customers_level_based", "PRICE"]]
agg_df.head()
```

```python
agg_df = agg_df.groupby("customers_level_based").agg({"PRICE": "mean"})
agg_df            
```

```python
agg_df = agg_df.reset_index()
agg_df.head()
```

```python
agg_df["SEGMENT"] = pd.qcut(agg_df["PRICE"], 4, labels=["D", "C", "B", "A"])
agg_df.head(30)
```
